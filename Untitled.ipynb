```python
from bs4 import BeautifulSoup
```

```python
with open("menu.php", 'rb') as menu:
    soup = BeautifulSoup(menu)
```

```python
deletes = soup(["script", "style", "!--"])
```

```python
print(deletes)
```

```python
for d in deletes:
    d.decompose()
```

```python
print(soup)
```

```python
menu_txt = soup.get_text()
```

```python
print(menu_txt)
```

```python
menu_txt_encoded = menu_txt.encode("utf-8", 'ignore')
```

```python
file = open("menu.txt", 'wb')
file.write(menu_txt_encoded)
file.close()
```
